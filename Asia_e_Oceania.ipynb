```r
ggplot(df_asia1)+
  aes(`Anos de governo`, fill = Ideologia)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+
  scale_fill_manual(values = cores_cat) +
  labs(title = 'Anos de governo por Ideologia')
```

```r
# Criando um gr√°fico de barras para vari√°veis categ√≥ricias
ggplot(df_asia1) + 
  aes(Ideologia, fill = Regime) + 
  geom_bar() + 
  theme_light() +
  scale_fill_manual(values = cores_rgb)+
  coord_flip() +
  labs(title = 'Ideologia de acordo com o regime')
```

A maior parte dos regimes presidencialistas com parlamento era declaradamente de esquerda.

```r
# An√°lise multifacetada
ggplot(df_asia1) +
  aes(x = `Anos de governo`, y = Religi√£o , color = Ideologia) +
  geom_point(alpha = 0.7, size = 3) +
  theme_light() +
  scale_color_manual(values = cores_cat) +
  facet_grid(Regime ~ `Restri√ß√µes √† reelei√ß√£o`)
```

Houveram governos islamicos de esquerda e governos hindus de direita. Tamb√©m houveram governos crist√£os de direita e de centro. No geral, religi√£o nao parece ser o definidor da dura√ß√£o do mandato na Asia e oceania.
