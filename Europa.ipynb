# Plots

```r
df_europa1 <- na.omit(df_europa)

ggplot(df_europa1)+
  aes(`Anos de governo`, fill = Regime)+ 
  geom_histogram(bins = 20, position = 'identity', alpha = 0.5)+
  theme_light()+
  scale_fill_manual(values = cores_cat) +
  labs(title = 'Anos de governo por Regime')
```

```r
# Gr√°fico de densidade de probabilidade
ggplot(df_europa1)+
  aes(`Anos de governo`, fill = Regime)+ 
  geom_density(position = 'identity', alpha = 0.5)+
  theme_light()+
  scale_fill_manual(values = cores_cat) +
  labs(title = 'Anos de governo por Regime')
```

A maior parte das experiencias parlamentristas durou ate 10 anos. As presidencialistas com parlamento foram as mais longas

```r
ggplot(df_europa1)+
  aes(`Anos de governo`, fill = Religi√£o)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+ 
  scale_fill_manual(values = cores_cat) +
  labs(title = 'Anos de governo por Religi√£o')
```

Os governos declaradamente cat√≥licos foram mais longevos. Os governos isl√°micos duraram mais tempo que crist√£os

```r
ggplot(df_europa1)+
  aes(`Anos de governo`, fill = Ideologia)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+
  scale_fill_manual(values = cores_cat) +
  labs(title = 'Anos de governo por Ideologia')
```

Os governos de esquerda foram os mais longos. Houveram mais governos de centro.

```r
ggplot(df_europa1)+
  aes(`Anos de governo`, fill = `Mandato finito`)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+ 
  theme_light()+
  scale_fill_manual(values = cores_rgb) +
  labs(title = 'Anos de governo x Regime nacionalista ou n√£o')
```

A maior parte dos mandatos "sem limita√ß√£o" duraram menos de 10 anos. Houveram relativamente mais governos de "mandato infinito" e curtas dura√ß√µes que mandatos mais longos. A maior parte dos governos sem restri√ß√µes a reelei√ß√£o duraram mais tempo.

```r
ggplot(df_europa1)+
  aes(`Anos de governo`, fill = Nacionalista)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+ 
  scale_fill_manual(values = cores_rgb) +
  labs('Anos de governo de regimes nacionalistas')
```

A maioria dos governos n√£o era nacionalista, e foram de maior dura√ß√£o. Os governos nacionalistas duraram no m√°ximo 15 anos.

```r
ggplot(df_europa1)+
  aes(`Anos de governo`, fill = `Restri√ß√µes √† reelei√ß√£o`)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+
  scale_fill_manual(values = cores_rgb) +
  labs(title = 'Anos de governo de acordo com restri√ß√µes √† reelei√ß√£o')
```

Hoveram alguns mandatos sem limita√ß√£o que duraram 3 d√©cadas. No geral a maioria dos governos sem restri√ß√£o chegou a uma d√©cada de mandato, possivelmente por conta das restri√ß√µes constitucionais.

```r
# Criando um gr√°fico de barras para vari√°veis categ√≥ricias
ggplot(df_europa1) + 
  aes(Ideologia, fill = Regime) + 
  geom_bar() + 
  theme_light() +
  scale_fill_manual(values = cores_rgb)+
  coord_flip() +
  labs(title = 'Ideologia de acordo com o regime')
```

Houveram mais experiencias parlamentaristas c presediente  de esquerda que de direita. N√£o houveram regimes presidencialistas com parlamento de centro na europa.

```r
ggplot(df_europa1) + 
  aes(`Restri√ß√µes √† reelei√ß√£o`, fill = Ideologia) + 
  geom_bar() + 
  theme_light() +
  scale_fill_manual(values = cores_cat)+
  coord_flip() +
  labs(title = 'Ideologia de governo x Restri√ß√µes √† reelei√ß√£o')
```

Houveram mais governos de esquerda em pa√≠ses com restri√ß√£o ate dois mandatos consecutivos. Houveram mais governos de direita em pa√≠ses sem limita√ß√µes a reelei√ß√£o


```r
# An√°lise multifacetada
ggplot(df_europa1) +
  aes(x = `Anos de governo`, y = Religi√£o , color = Ideologia) +
  geom_point(alpha = 0.7, size = 3) +
  theme_light() +
  scale_color_manual(values = cores_cat) +
  facet_grid(Regime ~ `Restri√ß√µes √† reelei√ß√£o`)
  labs(title = 'Regimes de acordo com o regime, nacionalismo, restri√ß√µes a reelei√ß√£o e ideologia')
```

N√£o hoveram experiencias de esquerda cat√≥licas, crist√¢s ou islamicas na europa. A maior parte das experiencias de direita era catolica ou crist√£o.
