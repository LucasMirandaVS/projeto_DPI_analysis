# Definindo a região

```r
df_Africa <- df1 %>%
  filter(País %in% c("Algeria" ,"Angola", "Benin", "Botswana",
                     "Burkina Faso", "Burundi", "C. Verde Is.",
                     "Cameroon", "Cent. Af. Rep.", "Chad",
                     "Comoro Is.", "Congo", "Congo (DRC)",
                     "Cote d'Ivoire", "Djibouti", "Egypt",
                     "Eq. Guinea", "Eritrea", "eSwatini",
                     "Ethiopia", "Gabon", "Gambia",
                     "Ghana", "Guinea", "Guinea-Bissau",
                     "Kenya", "Lesotho", "Liberia",
                     "Libya", "Madagascar", "Malawi",
                     "Mali", "Mauritania", "Mauritius",
                     "Morocco", "Mozambique", "Namibia",
                     "Niger", "Nigeria", "Rwanda",
                     "S. Africa", "Senegal", "Sierra Leone",
                     "Solomon Is.", "Somalia", "South Sudan",
                     "Sudan", "Swaziland", "Tanzania",
                     "Togo", "Tunisia","Uganda", 
                     "Zambia", "Zimbabwe"))
```

```r
df_africa1 <- na.omit(df_Africa)

ggplot(df_africa1)+
  aes(`Anos de governo`, fill = Regime)+ 
  geom_histogram(bins = 20, position = 'identity', alpha = 0.5)+
  theme_light()+
  scale_fill_manual(values = cores_cat) +
  labs(title = 'Anos de governo por Regime')
```

Maioria esmagadora é presidencialista. Os governos presidencialistas também foram os mais longos.

```r
# Gráfico de densidade de probabilidade
ggplot(df_africa1)+
  aes(`Anos de governo`, fill = Regime)+ 
  geom_density(position = 'identity', alpha = 0.5)+
  theme_light()+
  scale_fill_manual(values = cores_cat)  +
  labs(title = 'Anos de governo por Regime')
```

A maior parte das experiencias parlamentaristas durou ate 20 anos. As presidencialistas foram as mais longas

```r
ggplot(df_africa1)+
  aes(`Anos de governo`, fill = Religião)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+ 
  scale_fill_manual(values = cores_cat)  +
  labs(title = 'Anos de governo por Religião')
```

Os governos declaradamente Hindus foram mais longevos. Os governos islámicos duraram menos tempo, mas foram mais numerosos

```r
ggplot(df_africa1)+
  aes(`Anos de governo`, fill = Ideologia)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+
  scale_fill_manual(values = cores_cat)  +
  labs(title = 'Anos de governo por Ideologia')
```

Os governos de esquerda foram os mais longos. Houveram mais governos de centro

```r
ggplot(df_africa1)+
  aes(`Anos de governo`, fill = `Mandato finito`)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+ 
  theme_light()+
  scale_fill_manual(values = cores_rgb) +
  labs(title = 'Anos de Governo x Mandato finito')
```

Houveram relativamente mais governos de "mandato finito". A maior parte dos governos sem restrições a reeleição duraram mais tempo. Há uma quantidade considerável de governos sem limitações que duraram décadas.

```r
ggplot(df_africa1)+
  aes(`Anos de governo`, fill = Nacionalista)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+ 
  scale_fill_manual(values = cores_rgb) +
  labs('Anos de governo de regimes nacionalistas')
```

Ser nacionalista não influenciou muito na duração de governo. Houveram mais governos não nacionalistas.

```r
ggplot(df_africa1)+
  aes(`Anos de governo`, fill = `Restrições à reeleição`)+ 
  geom_density(position = 'identity', alpha = 0.5, color = NA)+
  theme_light()+
  scale_fill_manual(values = cores_cat) +
  labs(title = 'Anos de governo de acordo com restrições à reeleição')
```

Houveram alguns mandatos longos porém com limitações. Houve uma quantidade considerável de regimes com até 3 mandatos consecutivos.

```r
# Criando um gráfico de barras para variáveis categóricias
ggplot(df_africa1) + 
  aes(Ideologia, fill = Regime) + 
  geom_bar() + 
  theme_light() +
  scale_fill_manual(values = cores_rgb)+
  coord_flip() +
  labs(title = 'Ideologia vigente de acordo com o regime')
```

A maior parte dos regimes presidencialistas era declaradamente de esquerda. Houveram poucas experiencias parlamentaristas na África.

```r
ggplot(df_africa1) + 
  aes(`Restrições à reeleição`, fill = Ideologia) + 
  geom_bar() + 
  theme_light() +
  scale_fill_manual(values = cores_cat)+
  coord_flip() +
  labs(title = 'Ideologia de governo x Restrições à reeleição')
```

O unico regime com 3 mandatos consecutivos era de esquerda

```r
# Análise multifacetada
ggplot(df_africa1) +
  aes(x = `Anos de governo`, y = Religião , color = Ideologia) +
  geom_point(alpha = 0.7, size = 3) +
  theme_light() +
  scale_color_manual(values = cores_cat) +
  facet_grid(Regime ~ `Restrições à reeleição`)
  labs(title = 'Regimes de acordo com o regime, nacionalismo, restrições a reeleição e ideologia')
```

Como no caso da europa, não houveram regimes de esquerda declarademnete hindus ou islamicos.
